```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
qpcr_selection_databy_round = np.array([
    [148073555000000, 0, 0, 0, 1480735550],
    [78453090000, 46686230, 8597880, 2931700, 31416025],
    [7495310000, 5841170, 5415000, 6563930, 11173805],
    [1098286611500, 14049550, 1069130, 2149945, 48908185],
    [261814775000, 16806830, 8240015, 10686265, 907983590],
    [625598010000, 42260130, 23457330, 5439960, 6024870000]])

input_by_round = qpcr_selection_databy_round[:,:1]
selection_results_by_round = qpcr_selection_databy_round[:,1:]

selection_results_by_round_percentage = (
    selection_results_by_round / input_by_round) * 100

dummy_selection1 = selection_results_by_round_percentage[:,:1].flatten()
dummy_selection2 = selection_results_by_round_percentage[:,1:2].flatten()
dummy_selection3 = selection_results_by_round_percentage[:,2:3].flatten()
positive_selection = selection_results_by_round_percentage[:,3:4].flatten()
```

```python
dummy_selection2
```

```python
# Create 'ggplot' style
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(
            1, 1,
            figsize = (8, 6),
            dpi = 300)

n_groups = len(dummy_selection1)
index = np.arange(n_groups)

bar_width = 0.2

dummy1 = ax.bar(
    index,
    dummy_selection1,
    width=bar_width,
    color = '#4F4F4F',
    label='Dummy 1\nPHD2(-)'
    )

dummy2 = ax.bar(
    index + 1 * bar_width,
    dummy_selection2,
    width=bar_width,
    color='#8B8B8B',
    label='Dummy 2\nPHD2(-)'
    )

dummy3 = ax.bar(
    index + 2 * bar_width,
    dummy_selection3,
    width=bar_width,
    color='#C7C7C7',
    label='Dummy 3\nPHD2(-)'
    )

positive = ax.bar(
    index + 3 * bar_width,
    positive_selection,
    width=bar_width,
    color='#fc4f30',
    label='Target Protein\nPHD2(+)'
    )

legend_text = [
    'Dummy 1 PHD2(-)', 'Dummy 2 PHD2(-)',
    'Dummy 3 PHD2(-)', 'Target Protein PHD2(+)']

# ax.legend(
# #     legend_text,
#     loc = 'upper center',
#     bbox_to_anchor = (0.5, -0.10),
#     fancybox = True,
#     shadow = False,
#     fontsize = 10,
#     ncol = 4)

ax.tick_params(labelsize = 12)
ax.set_xlabel(
    'Selection Cycle (#)')
ax.set_ylabel(
    'Library Enrichment (%)')

ax.set_xticks(index + bar_width)
ax.set_xticklabels(['1', '2', '3', '4', '5', '6']) 
    
plt.legend()

qpcr_selection_results_png = 'qpcr_selection_results.png'

fig.savefig(
    qpcr_selection_results_png,
    bbox_inches = 'tight',
    dpi = 300)

# plt.tight_layout()
plt.show()
```
